##Hipoteza1
U proseku najvise diplomira studenata iz oblasti "poslovanje, administracija i pravo".

Da bismo ispitali postavljenu hipotezu potrebno je da prodjemo kroz nekoliko koraka, jer sagledavanje grafikona iznad moze nam se uciniti da je hipoteza zaista tacna.

```
dfOblast.where(dfOblast.nazivOblasti!="укупно").orderBy("vrednost", ascending=False).show(5, truncate=False)
```

S obzirom da nemamo najpregledniji uvid u podatke iznad, koristicemo pivot. Kolone ce nam sada biti godine i na taj nacin cemo moci da vidimo koliko je bilo diplomiranih studenata za svaku oblastu kroz godine iz dataseta.

```
pivotDF = dfOblast.where(dfOblast.nazivTer == "Republika Srbija").groupBy("nazivOblasti").pivot("godina").sum("vrednost")
pivotDF.orderBy("2016","2017","2019", ascending=False).show()
```

Ovo nije dovoljno da bismo mogli da potvrdimo hipotezu, pa podatke o diplomiranim studentima cemo uporediti sa podacima upisanih studenata u tim oblastima kroz godine. Radi bolje preglednosti i ovde cemo koristiti pivotiranje.

```
pivotDF1 = dfUpisani.where(dfUpisani.naziv_Ter == "Republika Srbija").groupBy("naziv_Oblasti").pivot("god").sum("brojUpisanih")
pivotDF1 = pivotDF1.withColumnRenamed("2016", "2016upisani")\
                 .withColumnRenamed("2017", "2017upisani")\
                 .withColumnRenamed("2018", "2018upisani")\
                 .withColumnRenamed("2019", "2019upisani")
pivotDF1.orderBy("2016upisani","2017upisani","2018upisani","2019upisani",ascending=False).show()
```

Vidimo se da najvise studenata i upise fakultet iz oblasti poslovanje, ali ni ovo nije dovoljno da bismo mogli bilo sta da tvrdimo. 
U sledecem koraku cemo spojiti ove dve tabele kako bismo imali podatke na jednom mestu.

```
pivotDF = pivotDF.join(pivotDF1, pivotDF1.naziv_Oblasti == pivotDF.nazivOblasti, "inner")\
                 .drop("2018upisani")
pivotDF.show(5)
```

Konacno, racunamo procenat studenata koji je diplomirao u odnosu na one koji su upisali fakultet u toj oblasti.

```
pivotDF = pivotDF.withColumn("proc2016", 100*col("2016")/col("2016upisani"))\
                 .withColumn("proc2017", 100*col("2017")/col("2017upisani"))\
                 .withColumn("proc2019", 100*col("2019")/col("2019upisani"))
```

Rezultate sortiramo kako bismo videli u kojoj oblasti studenti u proseku najvise diplomiraju.

```
pivotDF.orderBy("proc2016","proc2017","proc2019",ascending=False).show()
```

```
display(pivotDF.orderBy("proc2016","proc2017","proc2019",ascending=False))
```

Zakljucak: 2016. i 2017. u odnosu na broj upisanih studenata, najvise diplomiralo iz oblasti "poljoprivreda,sumarstvo, ribarstvo i veterina" dok je 2019. najvise diplomiralo studenata iz oblasti zdravstva. U preseku tih podataka, dobijamo da generalno najvise studenata diplomira iz oblasti poljoprivrede kada se ti podaci uporede sa upisanim studentima u toj oblasti.
Konacno, dobijamo da je nasa hipoteza netacna.

##Hipoteza 2
U oblasti "drustvene nauke, novinarstvo i informisanje" najvise studenata diplomira sa teritorije Republike Srbije - Sever.

Kao prvi korak, koristicemo pivot funkciju kako bismo imali pregledne podatke. Na taj nacin moci cemo da pratimo broj diplomiranih studenata po godinama na teritorijama Republike Srbije.

```
pivotDF = dfOblast.groupBy("nazivOblasti", "nazivTer").pivot("godina").sum("vrednost")
pivotDF.orderBy("nazivTer").show(10,truncate=False)
```

Izdvajamo podatke vezane samo za oblast koja nas zanima i sortiramo ih po broju diplomiranih kroz godine.

```
pivotDF = dfOblast.where(dfOblast.nazivOblasti == "друштвене науке, новинарство и информисање").groupBy("nazivTer").pivot("godina").sum("vrednost")
#na.fill zbog KiM
pivotDF.na.fill({"2016":0, "2017":0, "2019":0}).orderBy("2016","2017","2019", ascending=False).show(10,truncate=False)
```

Vidimo da je 2016. i 2017. godine najvise studenata diplomiralo u ovoj oblasti iz oblasti Sever, a 2019. godine u Regionu Juzne i Istocne Srbije, i Vojvodini.
Prikazujemo prosek diplomiranih studenata po oblastima.

```
display(dfOblast.na.fill({"vrednost":0}).where((dfOblast.nazivOblasti  == "друштвене науке, новинарство и информисање") & (dfOblast.nazivTer != "Republika Srbija")).groupBy("nazivTer").avg("vrednost").orderBy("nazivTer"))
```

Najvise studenata diplomira u Regionu Sever, ali potrebno je proveriti u odnosu na upisane studente u ovoj oblasti.

```
pivotDF = dfUpisani.where((dfUpisani.naziv_Oblasti == "друштвене науке, новинарство и информисање") &(dfUpisani.god != 2018)).groupBy("naziv_Ter").pivot("god").sum("brojUpisanih")
#na.fill zbog KiM
pivotDF.na.fill({"2016":0, "2017":0, "2019":0}).orderBy("naziv_Ter").orderBy("2016","2017","2019").show(10,truncate=False)
```

Prikazujemo prosek upisanih studenata iz oblasti drustvenih nauka po regionima Republike Srbije.

```
display(dfUpisani.na.fill({"brojUpisanih":0}).where((dfUpisani.naziv_Oblasti  == "друштвене науке, новинарство и информисање") & (dfUpisani.naziv_Ter != "Republika Srbija")).groupBy("naziv_Ter").avg("brojUpisanih").orderBy("naziv_Ter"))
```

Najvise studenata i upise fakultet u oblasti drustvenih nauka u Regionu Sever, ali ni to nije dovoljno da bismo potvrdili hipotezu. Racunacemo procenat diplomiranih u odnosu na upisane, a pre toga spojicemo ove dve tabele kako bi se podaci nalazili na jednom mestom.

```
cond = [dfOblast.nazivOblasti == dfUpisani.naziv_Oblasti, dfOblast.godina==dfUpisani.god, dfOblast.nazivTer==dfUpisani.naziv_Ter]
dfJoin = dfOblast.join(dfUpisani, cond, "inner")
dfJoin.show(5)
```

Sumiramo vrednosti kako bismo mogli da uradimo procenat za godine koje posmatramo.

```
dfJoin1 = dfJoin.groupBy("nazivTer","nazivOblasti").sum("brojUpisanih", "vrednost")
dfJoin1 = dfJoin1.withColumnRenamed("sum(brojUpisanih)", "upisaniUkupno")\
                 .withColumnRenamed("sum(vrednost)", "diplomiraniUkupno")
dfJoin1.orderBy("nazivTer").show(5)
```

Radimo procenat diplomiranih studenata u odnosu na upisane studente.

```
dfJoin1 = dfJoin1.withColumn("procenat diplomiranih", 100*col("diplomiraniUkupno")/col("upisaniUkupno"))
dfJoin1.orderBy("nazivTer").show(5)
```

Na kraju, prikazujemo procenat diplomiranih za oblast drustvenih nauka po regionima.

```
display(dfJoin1.na.fill({"procenat diplomiranih":0}).where((dfJoin1.nazivOblasti  == "друштвене науке, новинарство и информисање") & (dfJoin.nazivTer != "Republika Srbija")).groupBy("nazivTer").sum("procenat diplomiranih").orderBy("nazivTer"))
```

Rezultati pokazuju da najvise studenata sa teritorije Juzne i Istocne Srbije diplomira u oblasti drustvenih nauka.

##Streaming
Strimujemo podatke i grupisemo ih po godinama, a upit radimo za broj upisanih Z/M po godini.

```
inputPath = 'dbfs:/projekat/dfUpisaniPol.csv'

dfSchema = StructType([ StructField("idindikator", StringType(), True), StructField("IDTer", StringType(), True), StructField("nTer", StringType(), True), StructField("godina", IntegerType(), True), StructField("IDPol", IntegerType(), True), StructField("nPol", StringType(), True), StructField("brojUpisanih", IntegerType(), True) ])

staticInputDF = (
  spark
    .read
    .schema(dfSchema)
    .option("header", "true")
    .csv(inputPath)
)

staticInputDF = staticInputDF.where((staticInputDF.nTer == "Republika Srbija")&(staticInputDF.nPol != "Ukupno"))

display(staticInputDF.groupBy(staticInputDF.nPol, staticInputDF.godina).sum("brojUpisanih"))
```

```
staticSumDF = (
  staticInputDF
    .groupBy(staticInputDF.nPol, staticInputDF.godina, staticInputDF.brojUpisanih).sum("brojUpisanih")
)

staticSumDF.createOrReplaceTempView("static_sums")
```

```
%sql select nPol, godina, brojUpisanih from static_sums order by godina
```

```
streamingInputDF = (
  spark
    .readStream                       
    .schema(dfSchema)               
    .option("maxFilesPerTrigger", 1)  
    .csv(inputPath, header = True)
)

streamingInputDF = streamingInputDF.where((streamingInputDF.nTer == "Republika Srbija")&(streamingInputDF.nPol != "Ukupno"))

streamingCountsDF = (                 
  streamingInputDF.groupBy(streamingInputDF.godina, streamingInputDF.nPol, streamingInputDF.brojUpisanih).sum("brojUpisanih")
)

# Is this DF actually a streaming DF?
streamingCountsDF.isStreaming
```

```
spark.conf.set("spark.sql.shuffle.partitions", "2")  

query = (
  streamingCountsDF
    .writeStream
    .format("memory")        # memory = store in-memory table 
    .queryName("sums")     # sums = name of the in-memory table
    .outputMode("complete")  # complete = all the counts should be in the table
    .start()
)
```
